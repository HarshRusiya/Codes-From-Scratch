```python
import cv2
import matplotlib.pyplot as plt
from PIL import Image
from skimage import io
import numpy as np
import collections
from glob import glob
from scipy.io import wavfile
import math





def applyingwindow(wav,w_size):
    hop=160
    w=np.hanning(w_size) #hamming window
    feature_vector=np.zeros(([w_size,311]),dtype='float64')#since we have 300*160 +400 =311 points
    after_window=np.zeros((wav.shape))
    for i in range(0,len(wav)-w_size,hop):
        after_window[i:i+w_size]=wav[i:i+w_size]*w   #applying window
    after_window[len(wav)-w_size:len(wav)]=wav[len(wav)-w_size:len(wav)]*w   
    j=0
    for i in range(0,len(wav)-w_size,hop):
        feature_vector[:,j]=after_window[i:i+400]   
        j=j+1
    feature_vector[:,310]=after_window[len(wav)-w_size:len(wav)] 
    return feature_vector
def fourier(feature_vector_clean):
    data_matrix_clean=np.zeros([128,feature_vector_clean.shape[1]])
    for i in range(feature_vector_clean.shape[1]): 
        X_c=np.fft.fft((feature_vector_clean[:,i]),256)
        X_c=np.log(abs(X_c)+1)
        data_matrix_clean[:,i]=X_c[:128]
    return data_matrix_clean
def calcwhitening(data_matrix):
    S=np.cov(data_matrix,bias=True)          #Covariance Matrix
    eigenvalues,eigenvectors=np.linalg.eig(S)
    idx = eigenvalues.argsort()[::-1]   
    eigen = eigenvalues[idx] # lamda 
    eigenvec = eigenvectors[:,idx]
    L_sqrinv=np.zeros([128,128],dtype='float64')
    U=np.zeros([128,128],dtype='float64')
    for i in range(len(eigen)):    #Caluclating 1/sqrt(L)
        L_sqrinv[i,i]=1/(np.sqrt(eigen[i]))
        eigenvec[:,i]=eigenvec[:,i]/np.linalg.norm(eigenvec[:,i])
        U[:,i]=eigenvec[:,i]    
    whitening=(L_sqrinv@(U.T))
    return whitening    
def whitening(data_matrix):
    imgmean_c=np.mean(data_matrix_clean,axis=1).reshape(128,1)
    data_matrix_clean=(data_matrix_clean-imgmean_c)
    Yn_clean=whitening@data_matrix_clean
    return whitening

#PartA
f_name='clean.wav'
fs, wav=wavfile.read(f_name)
w_size=400    #got by 16000hz*25ms
hop=160       #got by 16000hz*10ms
feature_vectors=applyingwindow(wav,w_size)
data_matrix=fourier(feature_vectors)
whitened=calcwhitening(data_matrix)
Yn=whitened@data_matrix   
sampled_S=np.cov(Yn) 
total=0
for i in range(sampled_S.shape[0]):
    for j in range(sampled_S.shape[1]):
        if (i!=j):
            total+=abs(sampled_S[i][j])
avg=total/(np.square(128)-128)  
print("the average of the absolute value of the non-diagonal entries of the sampled covariance matrix of the “whitenned” clean is",avg)
f_name='noise.wav'
fs, wav=wavfile.read(f_name)
feature_vectors=applyingwindow(wav,w_size)
data_matrix=fourier(feature_vectors)
Yn=whitened@data_matrix
sampled_S=np.cov(Yn) 
total=0
for i in range(sampled_S.shape[0]):
    for j in range(sampled_S.shape[1]):
        if (i!=j):
            total+=abs(sampled_S[i][j])
avg=total/(np.square(128)-128)  
print("the average of the absolute value of the non-diagonal entries of the sampled covariance matrix of the “whitenned” noise image by using whitening of cleaned image is",avg)

#PartB
f_name='noisy.wav'
fs, wav=wavfile.read(f_name)
w_size=400    #got by 16000hz*25ms
hop=160       #got by 16000hz*10ms
feature_vectors=applyingwindow(wav,w_size)
data_matrix=fourier(feature_vectors)
whitened=calcwhitening(data_matrix)
Yn=whitened@data_matrix   
sampled_S=np.cov(Yn) 
total=0
for i in range(sampled_S.shape[0]):
    for j in range(sampled_S.shape[1]):
        if (i!=j):
            total+=abs(sampled_S[i][j])
avg=total/(np.square(128)-128)  
print("the average of the absolute value of the non-diagonal entries of the sampled covariance matrix of the “whitenned” noise is",avg)
f_name='clean.wav'
fs, wav=wavfile.read(f_name)
feature_vectors=applyingwindow(wav,w_size)
data_matrix=fourier(feature_vectors)
Yn=whitened@data_matrix
sampled_S=np.cov(Yn) 
total=0
for i in range(sampled_S.shape[0]):
    for j in range(sampled_S.shape[1]):
        if (i!=j):
            total+=abs(sampled_S[i][j])
avg=total/(np.square(128)-128)  
print("the average of the absolute value of the non-diagonal entries of the sampled covariance matrix of the “whitenned” clean image by using whitening of cleaned image is",avg)

#PartC 
#here we uses the same distribution clean-clean and noisy-noisy
#PartA
f_name='clean.wav'
fs, wav=wavfile.read(f_name)
w_size=400    #got by 16000hz*25ms
hop=160       #got by 16000hz*10ms
feature_vectors=applyingwindow(wav,w_size)
data_matrix=fourier(feature_vectors)
whitened=calcwhitening(data_matrix)
Yn=whitened@data_matrix   
sampled_S=np.cov(Yn) 
total=0
for i in range(sampled_S.shape[0]):
    for j in range(sampled_S.shape[1]):
        if (i!=j):
            total+=abs(sampled_S[i][j])
avg=total/(np.square(128)-128)  
print("the average of the absolute value of the non-diagonal entries of the sampled covariance matrix of the “whitenned” clean is",avg)
f_name='clean.wav'
fs, wav=wavfile.read(f_name)
feature_vectors=applyingwindow(wav,w_size)
data_matrix=fourier(feature_vectors)
Yn=whitened@data_matrix
sampled_S=np.cov(Yn) 
total=0
for i in range(sampled_S.shape[0]):
    for j in range(sampled_S.shape[1]):
        if (i!=j):
            total+=abs(sampled_S[i][j])
avg=total/(np.square(128)-128)  
print("the average of the absolute value of the non-diagonal entries of the sampled covariance matrix of the “whitenned” clean image by using whitening of cleaned image is",avg)

#PartA
f_name='noise.wav'
fs, wav=wavfile.read(f_name)
w_size=400    #got by 16000hz*25ms
hop=160       #got by 16000hz*10ms
feature_vectors=applyingwindow(wav,w_size)
data_matrix=fourier(feature_vectors)
whitened=calcwhitening(data_matrix)
Yn=whitened@data_matrix   
sampled_S=np.cov(Yn) 
total=0
for i in range(sampled_S.shape[0]):
    for j in range(sampled_S.shape[1]):
        if (i!=j):
            total+=abs(sampled_S[i][j])
avg=total/(np.square(128)-128)  
print("the average of the absolute value of the non-diagonal entries of the sampled covariance matrix of the “whitenned” noise is",avg)
f_name='noise.wav'
fs, wav=wavfile.read(f_name)
feature_vectors=applyingwindow(wav,w_size)
data_matrix=fourier(feature_vectors)
Yn=whitened@data_matrix
sampled_S=np.cov(Yn) 
total=0
for i in range(sampled_S.shape[0]):
    for j in range(sampled_S.shape[1]):
        if (i!=j):
            total+=abs(sampled_S[i][j])
avg=total/(np.square(128)-128)  
print("the average of the absolute value of the non-diagonal entries of the sampled covariance matrix of the “whitenned” noise image by using whitening of noise image is",avg)
```

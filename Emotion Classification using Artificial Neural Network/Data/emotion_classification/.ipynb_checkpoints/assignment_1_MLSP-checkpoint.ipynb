```python
import cv2
import matplotlib.pyplot as plt
from PIL import Image
from skimage import io
import numpy as np
import collections
np.set_printoptions(threshold=np.inf)

image1 = io.imread('subject01.happy.gif')
image2 = io.imread('subject02.happy.gif')
image3 = io.imread('subject02.sad.gif')
image4 = io.imread('subject03.sad.gif')
image5 = io.imread('subject04.happy.gif')
image6 = io.imread('subject04.sad.gif')
image7 = io.imread('subject05.sad.gif')
image8 = io.imread('subject06.happy.gif')
image9 = io.imread('subject06.sad.gif')
image10 = io.imread('subject07.happy.gif')
image11 = io.imread('subject07.sad.gif')
image12 = io.imread('subject09.happy.gif')
image13 = io.imread('subject09.sad.gif')
image14 = io.imread('subject10.happy.gif')
image15 = io.imread('subject10.sad.gif')
image16 = io.imread('subject11.sad.gif')
image17 = io.imread('subject12.happy.gif')
image18 = io.imread('subject12.sad.gif')
image19 = io.imread('subject13.happy.gif')
image20 = io.imread('subject13.sad.gif')
```

```python
img1=np.reshape(image1,(10201))
img2=np.reshape(image2,(10201))
img3=np.reshape(image3,(10201))
img4=np.reshape(image4,(10201))
img5=np.reshape(image5,(10201))
img6=np.reshape(image6,(10201))
img7=np.reshape(image7,(10201))
img8=np.reshape(image8,(10201))
img9=np.reshape(image9,(10201))
img10=np.reshape(image10,(10201))
img11=np.reshape(image11,(10201))
img12=np.reshape(image12,(10201))
img13=np.reshape(image13,(10201))
img14=np.reshape(image14,(10201))
img15=np.reshape(image15,(10201))
img16=np.reshape(image16,(10201))
img17=np.reshape(image17,(10201))
img18=np.reshape(image18,(10201))
img19=np.reshape(image19,(10201))
img20=np.reshape(image20,(10201))
```

```python
matrix1=np.array([img1,img2,img3,img4,img5,img6,img7,img8,img9,img10,img11,img12,img13,img14,img15,img16,img17,img18,img19,img20])
```

```python
matrix1
```

```python
X=np.zeros([20,10201])
for i in range(10201):
    X[:,i]=matrix1[:,i]-np.mean(matrix1[:,i]) 
```

```python
S=(X@np.transpose(X))
```

```python
eigenvalue,eigenvectors=np.linalg.eigh(S)
```

```python
eigenvalue
```

```python
eigenvectors
```

```python
new_dict = {k: v for k, v in zip(eigenvalue, eigenvectors)}  
r=collections.OrderedDict(sorted(new_dict.items()))
r
```

```python
eig=list(r.keys())
eigenv=list(r.values())
```

```python
k=np.zeros([20])
for j in range(-1,19):
    U=np.zeros([10201,20])
    for i in range(19,j,-1):
        U=np.transpose(X)*eig[i]/(np.sqrt(20*abs(eigenv[i])))
    Y1=np.transpose(U)@np.transpose(X[0,:])
    Y2=np.transpose(U)@np.transpose(X[1,:])
    Y3=np.transpose(U)@np.transpose(X[2,:])
    Y4=np.transpose(U)@np.transpose(X[3,:])
    Y5=np.transpose(U)@np.transpose(X[4,:])
    Y6=np.transpose(U)@np.transpose(X[5,:])
    Y7=np.transpose(U)@np.transpose(X[6,:])
    Y8=np.transpose(U)@np.transpose(X[7,:])
    Y9=np.transpose(U)@np.transpose(X[8,:])
    Y10=np.transpose(U)@np.transpose(X[9,:])
    Y11=np.transpose(U)@np.transpose(X[10,:])
    Y12=np.transpose(U)@np.transpose(X[11,:])
    Y13=np.transpose(U)@np.transpose(X[12,:])
    Y14=np.transpose(U)@np.transpose(X[13,:])
    Y15=np.transpose(U)@np.transpose(X[14,:])
    Y16=np.transpose(U)@np.transpose(X[15,:])
    Y17=np.transpose(U)@np.transpose(X[16,:])
    Y18=np.transpose(U)@np.transpose(X[17,:])
    Y19=np.transpose(U)@np.transpose(X[18,:])
    Y20=np.transpose(U)@np.transpose(X[19,:])
    y_0=np.array([Y1,Y2,Y5,Y8,Y10,Y12,Y14,Y17,Y19])
    y_1=np.array([Y3,Y4,Y6,Y7,Y9,Y11,Y13,Y15,Y16,Y18])
    y_mean0=y_0.mean()
    y_var0=y_0.var()
    y_mean1=y_1.mean()
    y_var1=y_1.var()
    J=np.square(y_mean1-y_mean0)/(np.square(y_var0)+np.square(y_var1))
    k[j]=J
```

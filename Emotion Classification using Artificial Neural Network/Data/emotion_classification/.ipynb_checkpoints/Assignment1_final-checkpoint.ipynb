```python
Data=np.transpose(matrix1)
Data.shape
```

```python
X=np.zeros([20,10201])
for i in range(10201):
    X[:,i]=Data[:,i]-np.mean(Data[:,i])  
X
```

```python
S=(X@np.transpose(X))/20
```

```python
eigenvalues,eigenvectors=np.linalg.eig(S)
```
